Complete all the Task.

• Convert the grayscale image into an RGB image where the grayscale values are replicated

• Display the converted RGB image using matplotlib.

```python
import numpy as np
import matplotlib.pyplot as plt
from PIL import Image
import requests
from io import BytesIO

# 1. Load and display a grayscale image
img_url = "https://plus.unsplash.com/premium_photo-1747054586711-4e5795a7f33b?q=80&w=774&auto=format&fit=crop&ixlib=rb-4.1.0&ixid=M3wxMjA3fDB8MHxwaG90by1wYWdlfHx8fGVufDB8fHx8fA%3D%3D"
try:
    response = requests.get(img_url, timeout=10)
    response.raise_for_status()
    gray_img = Image.open(BytesIO(response.content)).convert('L')

    print("1. Grayscale Image:")
    plt.imshow(gray_img, cmap='gray')
    plt.title("Grayscale Image")
    plt.axis('off')
    plt.show()

    # 2. Extract and display the middle section (150 pixels)
    gray_np = np.array(gray_img)
    h, w = gray_np.shape
    center_h, center_w = h // 2, w // 2
    crop_size = 150
    h_start, h_end = center_h - crop_size // 2, center_h + crop_size // 2
    w_start, w_end = center_w - crop_size // 2, center_w + crop_size // 2
    middle_section = gray_np[h_start:h_end, w_start:w_end]

    print("2. Middle Section (150px):")
    plt.imshow(middle_section, cmap='gray')
    plt.title("150x150 Center Crop")
    plt.axis('off')
    plt.show()

    # 3. Apply a simple threshold (binary image)
    binary_img = np.where(gray_np < 100, 0, 255).astype(np.uint8)

    print("3. Binary Image (Threshold: 100):")
    plt.imshow(binary_img, cmap='gray')
    plt.title("Binary Image")
    plt.axis('off')
    plt.show()

    # 4. Rotate the image 90 degrees clockwise
    rotated_img = gray_img.rotate(-90)

    print("4. Rotated Image (90 degrees clockwise):")
    plt.imshow(rotated_img, cmap='gray')
    plt.title("90 Deg Clockwise")
    plt.axis('off')
    plt.show()

    # 5. Convert grayscale image to an RGB image
    rgb_from_gray = np.stack([gray_np, gray_np, gray_np], axis=-1)

    print("5. Converted RGB Image:")
    plt.imshow(rgb_from_gray)
    plt.title("Grayscale to RGB")
    plt.axis('off')
    plt.show()
    print(f"New shape: {rgb_from_gray.shape}")

except Exception as e:
    print(f"Error processing image: {e}")
```

# 3 Image Compression and Decompression using PCA.
In this exercise, build a PCA from scratch using explained variance method for image compression task.
You are expected to compute the necessary matrices from the scratch. Dataset: Use image of your
choice.
1. Load and Prepare Data:
• Fetch an image of you choice.{If colour convert to grayscale}
• Center the dataset - Standaridze the Data.
• Calculate the covaraince matrix of the Standaridze data.
2. Eigen Decomposition and Identifying Principal Components:
• Compute Eigen Values and Eigen Vectors.
• Sort the eigenvalues in descending order and choose the top k eigenvectors corresponding to
the highest eigenvalues.
• Identify the Principal Components with the help of cumulative Sum plot.
3. Reconstruction and Experiment:

• Reconstruction: Transform the original data by multiplying it with the selected eigenvec-
tors(PCs) to obtain a lower-dimensional representation.

• Experiments: Pick Four different combination of principal components with various ex-
plained variance value and compare the result.

• Display the Results and Evaluate.

```python
import numpy as np
import matplotlib.pyplot as plt
from PIL import Image
import requests
from io import BytesIO

# 1. Load and Prepare Data
img_url = "https://plus.unsplash.com/premium_photo-1747054586711-4e5795a7f33b?q=80&w=774&auto=format&fit=crop&ixlib=rb-4.1.0&ixid=M3wxMjA3fDB8MHxwaG90by1wYWdlfHx8fGVufDB8fHx8fA%3D%3D"
response = requests.get(img_url)
img = Image.open(BytesIO(response.content)).convert('L')
img_data = np.array(img, dtype=np.float64)

# Standardize/Center the data (Subtract mean of each column)
mean_vec = np.mean(img_data, axis=0)
centered_data = img_data - mean_vec

# Calculate Covariance Matrix
cov_matrix = np.cov(centered_data, rowvar=False)

# 2. Eigen Decomposition
eigen_values, eigen_vectors = np.linalg.eigh(cov_matrix)

# Sort in descending order
sorted_idx = np.argsort(eigen_values)[::-1]
eigen_values = eigen_values[sorted_idx]
eigen_vectors = eigen_vectors[:, sorted_idx]

# Explained Variance and Cumulative Sum
explained_var = eigen_values / np.sum(eigen_values)
cumulative_var = np.cumsum(explained_var)

plt.figure(figsize=(8, 4))
plt.plot(cumulative_var, marker='o', linestyle='--')
plt.title("Cumulative Sum of Explained Variance")
plt.xlabel("Number of Principal Components")
plt.ylabel("Cumulative Variance")
plt.grid()
plt.show()

# 3. Reconstruction and Experiment
def reconstruct_image(data, eigenvectors, k, mean):
    top_k_vectors = eigenvectors[:, :k]
    low_dim_data = np.dot(data, top_k_vectors)
    reconstructed = np.dot(low_dim_data, top_k_vectors.T) + mean
    return reconstructed

# Experiment with 4 different k values
components = [5, 20, 50, 150]
fig, axes = plt.subplots(1, 4, figsize=(20, 5))

for i, k in enumerate(components):
    reconstructed_img = reconstruct_image(centered_data, eigen_vectors, k, mean_vec)
    var_val = cumulative_var[k-1] * 100
    axes[i].imshow(reconstructed_img, cmap='gray')
    axes[i].set_title(f"k={k}\nVar: {var_val:.2f}%")
    axes[i].axis('off')

plt.suptitle("PCA Reconstruction Experiments", fontsize=16)
plt.show()
```
